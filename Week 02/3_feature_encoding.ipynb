#### 1.Import Dependencies

```python
import os
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

| Variable Type | Preferred Encoding | Why?                                                                 |
|---------------|--------------------|----------------------------------------------------------------------|
| Nominal       | One-Hot Encoding   | No inherent order → avoids implying false ordinal relationships      |
| Ordinal       | Label Encoding     | Preserves order → small integers represent increasing levels         |


```python
df = pd.read_csv("../Week 01/Data/processed/churnModelling_binning_applied.csv")
df.head(4)
```

- Gender --> Nominal
- Geography --> Nominal
- CreditScoreBins --> Ordinal

#### 2. Encode Nominal Variables

```python
nominal_variables = ["Geography", "Gender"]

geography_dummies = pd.get_dummies(df['Geography'], prefix="Geography")
gender_dummies = pd.get_dummies(df['Gender'], prefix="Gender")
df_encoded = pd.concat([df, gender_dummies, geography_dummies], axis=1)
del df_encoded["Gender"]
del df_encoded["Geography"]
df_encoded
```

#### 3.Encode Ordinal Variables

```python
encode_dict_creditscore = {'Poor' : 0,
                           'Fair' : 1,
                           'Good' : 2,
                           'Very Good': 3,
                           'Excellent' : 4}

df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_creditscore)
df_encoded.head(10)
```

```python
df_encoded.to_csv("../Week 01/Data/processed/churnModelling_feature_encoded.csv")
```
